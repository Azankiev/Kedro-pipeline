```python
import pandas as pd
```

```python
df = pd.read_csv('../data/01_raw/RNP - ecsv_6_2023.csv', dtype={'PayerAccountId': str, 'LinkedAccountId': str})
df
```

```python
df_payer = df[['PayerAccountId', 'PayerAccountName']].copy()
df_payer = df_payer.drop_duplicates()
df_payer = df_payer.rename(columns={'PayerAccountId': 'Account ID', 'PayerAccountName': 'Name'})
df_payer
```

```python
df_linked = df[['LinkedAccountId', 'LinkedAccountName']].copy()
df_linked = df_linked.drop_duplicates().dropna()
df_linked = df_linked.rename(columns={'LinkedAccountId': 'Account ID', 'LinkedAccountName': 'Name'})
df_linked
```

```python
df_accounts_per_org = pd.concat([df_payer, df_linked]).drop_duplicates()
df_accounts_per_org
```

```python
df_accounts_per_org.to_csv('../data/01_raw/accounts_per_organization.csv', mode='a', header=False, index=False)
```
