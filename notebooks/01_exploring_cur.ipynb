```python
def s3_prefix_formatter(s3_key: str) -> str:
    year, month, fname = s3_key.split('/')
    year = year.replace('year=', '').zfill(4)
    month = month.replace('month=', '').zfill(2)
    fname = fname.replace('.snappy.parquet', '.csv')
    
    return f'{year}_{month}_{fname}'
```
